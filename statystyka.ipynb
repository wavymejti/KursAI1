```python
list_of_numbers = [6, 10, 3, 7, 12]

def policz_srednia(list_liczb):
  sum = 0
  for numer in list_liczb:
    sum += numer

  return sum / len(list_liczb)

srednia = policz_srednia(list_of_numbers)
print(srednia)
```

```python
lista1 = [1,2,3]
lista2 = [1,2,3,4,5,6]

def policz_mediane(lista):
  lista.sort()
  n = len(lista)

  if n % 2 == 0:
    return (lista[n // 2 - 1] + lista[n // 2]) / 2
  else:
    return lista[n // 2]

mediana = policz_mediane(lista1)
print(mediana)
mediana2 = policz_mediane(lista2)
print(mediana2)
```

```python
lista = [1,2,3,4,5,6,6,6,6,6]

def oblicz_mode(jakas_lista):
  ilosc_wystapien = {i : jakas_lista.count(i) for i in jakas_lista}
  return max(ilosc_wystapien, key = ilosc_wystapien.get)

print(oblicz_mode(lista))
```

```python
pajaki = [48, 49, 51, 52]
ender_dragony = [0, 0, 100, 100]
szkielety = [10, 30, 70, 90]
zombie = [49, 50, 50, 51]

import statistics
import matplotlib.pyplot as plt

def oblicz_wariancje(lista):
  srednia = statistics.mean(lista)
  wariancja = sum((x - srednia) ** 2 for x in lista) / len(lista)
  return wariancja, srednia

grupy = {
    "pajaki": pajaki,
    "ender_dragony": ender_dragony,
    "szkielety": szkielety,
    "zombie" : zombie
}

wariancja = {}
srednia = {}

for name, group in grupy.items():
  var, mean = oblicz_wariancje(group)
  wariancja[name] = var
  srednia[name] = mean
  print(f'{name} - srednia : {mean}, wariancjia: {var}')

#tworzenie wykresu
plt.figure(figsize=(10,6))
markers = ['o', 'x', 'D', '^']
colors = ['b', 'r', 'g', 'y']

for i, (name, group) in enumerate(grupy.items()):
  x_values = list(range(len(group)))
  y_value = [srednia[name]] * len(grupy)

  plt.scatter(x_values, group, color = colors[i], marker = markers[i], label = 'dane rzeczywiste')
  plt.plot(x_values, y_value, color=colors[i], linestyle = 'dashed')
```
