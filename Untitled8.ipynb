```python
# Generowanie histogramów z KDE – etykiety po angielsku,
# komentarze po polsku
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from scipy.stats import gaussian_kde
# Załaduj dane Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target,
iris.target_names)
# Podział 70 % train / 30 % test
df_train, df_test = train_test_split(df, test_size=0.30,
random_state=42)
# Lista cech

features = [
"sepal length (cm)",
"sepal width (cm)",
"petal length (cm)",
"petal width (cm)"
]
species = iris.target_names
# Rysowanie histogramów + KDE dla zbioru treningowego
plt.figure(figsize=(16, 12))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(
        data=df_train,
        x=feature,
        hue="species",
        kde=True
    )
    plt.title(f"Histogram of {feature} (train 70%)") #
    # tytuł po angielsku
    plt.xlabel(feature) # oś
    # X po angielsku
    plt.ylabel("Count") # oś Y
    # po angielsku
plt.tight_layout()
plt.show()
# Funkcja pomocnicza: wyznacz początek, wierzchołek i koniec
# KDE
def analyze_kde(values):
    kde = gaussian_kde(values)
    x_eval = np.linspace(values.min(), values.max(), 500)
    y_eval = kde(x_eval)
    peak = x_eval[np.argmax(y_eval)]
    return x_eval[0], peak, x_eval[-1]

# Zbieranie statystyk KDE dla pełnych danych
results = {}
for feature in features:
    results[feature] = {}
    for specie in species:
        vals = df[df['species'] == specie][feature].values
        start, peak, end = analyze_kde(vals)
        results[feature][specie] = {"start": start, "peak":
        peak, "end": end}
# Tabela podsumowująca
summary_df = pd.DataFrame([
    {"Feature": f, "Species": s, **vals}
    for f, d in results.items() for s, vals in d.items()
])
print(summary_df)
```
