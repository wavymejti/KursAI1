match wartosc:

  case wzorzec1:

  *kod do wykonania, jeśli wartość pasuje do
wzorca1*

  case wzorzec2:

  *kod do wykonania, jeśli wartość pasuje do
 wzorca2*
  case _:

  kod domyślny (jak "default" w switch)`

```python
wartosc = int(input("Podaj wartość: "))

match wartosc:
  case 1:
    print("wartosc równa 1")
  case 2:
    print("wartosc równa 2")
  case 3:
    print("wartosc równa 3")
  case _:
    print("wartosc nie pasuje do żadnego wzorca")
```

```python
owoce = ["jabłko", "banan", "kiwi", "pomidor", "malina", "mango", "marakuja", "pomarańcza"]

#wyswietlanie po indzieksie od poczatku
print(owoce[0])
print(owoce[1])
print(owoce[7])

#wyswietlanie po indeksie od konca
print(owoce[-1])
print(owoce[-8])

#chcemy fragment listy
#bierzemy elementy od indeksu 0 do 1
fragemnt = owoce[-3:7]
print(fragemnt)

#metody i funkcje przydatne w listach

# .append(element) dodaje nowy element na sam koniec listy
owoce.append("gruszka")
print(owoce)

# .pop(indeks) usuwa i zwraca element z podanego indeksu.
# (jesli nie podamy indesku to pop usunie ostatnia wartosc)
owoce.pop(8)
print(owoce)

# len(lista) - (od słowa length) czyli zwraca całkowitą liczbę elementów na liście
print(len(owoce))
```
