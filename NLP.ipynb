```python
#biblioteka NLTK
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt_tab')

text = 'NLP to skrót od Natural Language Processing, po polsku to przetwarzanie języka naturalnego'
print("Zdania", sent_tokenize(text))
print("Słowa", word_tokenize(text))
```

```python
from nltk.stem import WordNetLemmatizer, SnowballStemmer
lemmetizer = WordNetLemmatizer()
stemmer = SnowballStemmer("english")
nltk.download('wordnet')

words = ["running", "ran", "easily", "easy", "studied", "studying", "studies"]
print("{:<10} {:<10} {:<10}".format("Word", "Lemma", "Stemmer"))
for word in words:
    print("{:<10} {:<10} {:<10}".format(word, lemmetizer.lemmatize(word, pos = 'v'), stemmer.stem(word)))
```
