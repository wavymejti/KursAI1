```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
x = np.linspace(0, 100, 1000)
# Funkcje przynależności
trimf = fuzz.trimf(x, [20, 50, 80])
trapmf = fuzz.trapmf(x, [10, 30, 70, 90])
gaussmf = fuzz.gaussmf(x, mean=50, sigma=15)
plt.figure(figsize=(12, 6))
plt.plot(x, trimf, label='Trójkątna (20, 50, 80)', color='orange')
plt.plot(x, trapmf, label='Trapezowa (10, 30, 70, 90)',
color='green')
plt.plot(x, gaussmf, label='Gaussowska (μ=50, σ=15)', color='blue')
plt.title('Funkcje przynależności – scikit-fuzzy')
plt.xlabel('x')
plt.ylabel('Przynależność μ(x)')

plt.grid(True)
plt.legend()
plt.ylim(0, 1.1)
plt.show()
```
