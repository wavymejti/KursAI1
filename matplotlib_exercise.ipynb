```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
punkty_x = [-3,-2,-1,0,1,2,3]
punkty_y = []
for x in range(-3,4,1):
  wynik = 2 * x - 1
  punkty_y.append(wynik)
print(punkty_y)

plt.plot(punkty_x,punkty_y, c='r', marker='o', linestyle='--')
plt.grid()
plt.show()
```
