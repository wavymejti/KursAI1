```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
punkty_x = [-3,-2,-1,0,1,2,3]
punkty_y = []
for x in range(-3,4,1):
  wynik = 2 * x - 1
  punkty_y.append(wynik)
print(punkty_y)

plt.plot(punkty_x,punkty_y, c='r', marker='o', linestyle='--', label="y = 2x - 1")
plt.title("Wykres funkcji liniowej")
plt.legend()
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.grid()
plt.show()
```

```python
xpoints = np.linspace(-5,5,1000)
ypoints = 2 * np.square(xpoints) - 3
plt.plot(xpoints,ypoints, label = 'y = 2x^2 - 3')
plt.title("Wykres funkcji kwadratowej")
plt.legend()
plt.legend()
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.grid()
plt.show()
```

```python
x1 = np.linspace(-30, -10, 100)
x2 = np.linspace(-10,10, 100)
x3 = np.linspace(10,30, 100)

y1 = 5*x1 + 131
y2 = x2**2 - 19
y3 = (x3 - 9)**2 + 81

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.show()
```
